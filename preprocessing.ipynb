```python
import pandas as pd
```

```python
data = pd.read_json('generated_scores/generated_random_params_big_scores.jsonl', orient='records', lines=True)
```

```python
data.head()
```

```python
data = pd.concat([data.drop(['params'], axis=1), data['params'].apply(pd.Series)], axis=1)
```

```python
data.head()
```

```python
data = data.round({'top_p': 2, 'typical_p': 2, 'temperature': 2})
```

```python
top_k_range = [5, 25, 50, 75, 100, 150, 250, 350, 500, 750, 1000]

top_p_range = np.arange(0.05, 1.01, 0.05)

typical_p_range = top_p_range

temperature_range = np.arange(0.05, 1.21, 0.05)

change_type_range = ['top_p', 'typical_p', 'both']
```

```python
change_type_vocab = {'top_p': 0, 'typical_p': 1, 'both': 2}
top_p_vocab = {round(value, 2): i for value, i in zip(top_p_range, range(len(top_p_range)))}
typical_p_vocab = {round(value, 2): i for value, i in zip(typical_p_range, range(len(typical_p_range)))}
top_k_vocab = {round(value, 2): i for value, i in zip(top_k_range, range(len(top_k_range)))}
temperature_vocab = {round(value, 2): i for value, i in zip(temperature_range, range(len(temperature_range)))}
```

```python
data['change_type'] = data['change_type'].map(change_type_vocab)
data['top_p'] = data['top_p'].map(top_p_vocab)
data['top_k'] = data['top_k'].map(top_k_vocab)
data['typical_p'] = data['typical_p'].map(typical_p_vocab)
data['temperature'] = data['temperature'].map(temperature_vocab)
```

```python
data.head()
```

```python
columns = ['top_p', 'top_k', 'typical_p', 'temperature', 'do_sample']
data['params'] = data[columns].to_dict(orient='records')
```

```python
data = data.drop(columns=columns)
```

```python
data['number_of_answers'] = data['predicted_answers'].apply(lambda x: len(set([ans["answer"] for ans in x])))
```

```python
data_good = data[(data.mean_score >= data.mean_score.quantile(0.5)) & (data.number_of_answers >= 3) & (data.std_score >= 0.03)]
```
